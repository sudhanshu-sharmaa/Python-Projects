### Step 9. Which was the most-ordered item?

```python
# Group by item_name and sum the quantity
item_counts = chipo.groupby('item_name')['quantity'].sum()
# Sort in descending order
most_ordered = item_counts.sort_values(ascending=False)

# Display the top item
print("Most-ordered item:")
print(most_ordered.head(1))
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
# Group by item_name and sum the quantity
item_counts = chipo.groupby('item_name')['quantity'].sum()

# Find the most-ordered item and its total quantity
most_ordered_item = item_counts.idxmax()
most_ordered_quantity = item_counts.max()

# Display the result
print(f"The most-ordered item is '{most_ordered_item}' with {most_ordered_quantity} items ordered.")
```

### Step 11. What was the most ordered item in the choice_description column?

```python
item_counts = chipo.groupby('choice_description')['quantity'].sum()

# Find the most-ordered item and its total quantity
most_ordered_item = item_counts.idxmax()
most_ordered_quantity = item_counts.max()

# Display the result
print(f"The most-ordered item is '{most_ordered_item}' with {most_ordered_quantity} items ordered.")
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("Total revenue:", revenue)
```

### Step 15. How many orders were made in the period?

```python
total_orders = chipo['order_id'].nunique()
print("Total number of orders:", total_orders)
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1

# Total revenue
total_revenue = (chipo['quantity'] * chipo['item_price']).sum()

# Total number of orders
total_orders = chipo['order_id'].nunique()

# Average revenue per order
average_revenue = total_revenue / total_orders
print("Average revenue per order: ${:.2f}".format(average_revenue))

```
